**Dot Plot**

    Representation

     - Graphical representation of data using dots.

    
    Types
      
    - Wikinson Dot Plot
    
    - Cleveland Dot Plot

    Cleveland Dot Plot

    - created by William S.Cleveland.
    - Use position to plot the value.
    - similar to bar graph.

    Wikinson Dot Plot

    - Created by Leland Wikinson.
    - prevent the dots on the plot from overlapping
    - Displays a distribution of continuous data points.
    - Similar to histogram, but it display individual data points instead of bins.

    Application

    - To display the distribution of data
    - To cluster the value
    - To find outliers

    Understanding Data

    -  visualize the shape and spread of data
    -  comparing frequency distributions - indicates how often values in a dataset occurs.

    Clustering

    - group the data into different cluster

    Outliers

    - Identify any abnormal data
    
    Interpreting the Dot Plot
    - What is the minimum duration
    - What is the maximum duration

Styling Lines and Markers

The plt.plot function supports many arguments for styling lines and markers:

    color or c: Set the color of the line (supported colors)
    linestyle or ls: Choose between a solid or dashed line
    linewidth or lw: Set the width of a line
    markersize or ms: Set the size of markers
    markeredgecolor or mec: Set the edge color for markers
    markeredgewidth or mew: Set the edge width for markers
    markerfacecolor or mfc: Set the fill color for markers
    alpha: Opacity of the plot

Area Graph

Understanding Area Plot

               - It is similar to line graph
               - It is an extension of Line graph
               - It is obtained by filling the region between the line chart and axes with color
               - use different colors when more than one area plot
               - commonly emphasized with colors, textures
               - Used to track changes over time for one or more groups

          
Applications

               - compare a quantitative progress over time



Types

        - Overlapped Area chart
        - Stacked Area chart
        - Percentage based Area chart

```python
import matplotlib.pyplot as plt
```

```python
samosa =   [56,87,90,95,120,34]
days = ["Mon","Tue","Wed","Thurs","Fri","Sat"]

# color marker line
# Shorhand on formatting

plt.fill_between(days,samosa,color="pink")

plt.plot(days, samosa, color="blue",
         alpha=0.6, linewidth=2)

plt.title("Number of Samosa sold per day")

plt.xlabel("Days")
plt.ylabel("No.Samosa sold")




plt.show()
```

**Box Plot**

A box plot is a way of statistically representing the distribution of the data through five main dimensions:

Minimun: Smallest number in the dataset.
First quartile: Middle number between the minimum and the median.
Second quartile (Median): Middle number of the (sorted) dataset.
Third quartile: Middle number between median and maximum.
Maximum: Highest number in the dataset.

**Subplots**

To visualize multiple plots together, we can create a figure (overall canvas) and divide it into subplots, each containing a plot. With subplots, we usually work with the artist layer instead of the scripting layer.

Typical syntax is :

fig = plt.figure() # create figure
    ax = fig.add_subplot(nrows, ncols, plot_number) # create subplots
Where

nrows and ncols are used to notionally split the figure into (nrows * ncols) sub-axes,
plot_number is used to identify the particular subplot that this function is to create within the notional grid. plot_number starts at 1, increments across rows first and has a maximum of nrows * ncols as shown below.

```python
# Can macronutrients and calories affect ratings?
fig = plt.figure()

plt.rcParams['figure.figsize'] = [10,10]
plt.rcParams["font.weight"] = "bold"

plt.subplots_adjust(wspace=0.2, hspace=0.2)

fig.suptitle('Can calories and macronutrients affect cereal rating?', fontsize=25,fontweight="bold", color="black",
             position=(0.5,1.01))

ax1 = fig.add_subplot(221)
ax1.scatter('calories', 'rating', data= df, c="green")
ax1.set_title('Calories', color="green")


ax2 = fig.add_subplot(222)
ax2.scatter('fat', 'rating', data=df, c="orange")
ax2.set_title('Fat',  color="orange")


ax3 = fig.add_subplot(223)
ax3.scatter('protein', 'rating', data=df, c="brown")
ax3.set_title('Protein',  color="brown")

ax4 = fig.add_subplot(224)
ax4.scatter('carbo', 'rating', data=df, c="blue")
ax4.set_title("Carbs",  color="blue")


# We can see that as the amount of calories per serving, fats and carbs increase, the rating of the cereal decreases.
#We also see a positive relationship between protein and ratings.
# Can micronutrients affect ratings?
```
